# 折れ線グラフ

```python
x = np.random.randn(100)

fig, ax = plt.subplots()
ax.plot(x)
```

# 散布図

```python
x = np.random.random(100)
y = np.random.random(100)
s = 500 * np.random.random(100)

fig, ax = plt.subplots()
ax.scatter(x, y, s=s, alpha=0.5)
```

# ヒストグラム

```python
x = np.random.randn(1000)

fig, ax = plt.subplots()
h = ax.hist(x, bins='auto', alpha=0.5)
```

```python
fig, ax = plt.subplots()
h = ax.hist(x, bins='auto', histtype='step')
```

# 画像データを描画

```python
image = np.random.random((128, 128))

fig, ax = plt.subplots()
ax.imshow(image)
```

# データ範囲を指定

```python
x = np.random.randn(120)

fig, ax = plt.subplots()
ax.plot(x)
ax.set_xlim(0, 100)
```

```python
x = np.random.randn(120)

fig, ax = plt.subplots()
ax.plot(x)
ax.set_ylim(top=6)
```

```python
fig, ax = plt.subplots()
ax.plot(x)
ax.margins(x=0.10, y=0.10)
```

# ラベル、タイトル追加

```python
x = np.random.randn(120)

fig, ax = plt.subplots()
ax.plot(x)
ax.set(xlabel='Time(sec)', ylabel='Voltage(mv)', title='Awesome sensor data')
```

# 凡例追加

```python
x = np.random.randn(120)

fig, ax = plt.subplots()
ax.plot(x, label='legend')
ax.legend()
```

```python
x = np.linspace(0, 14, 100)
data = [np.sin(x + i * 0.5) * (13 - i) for i in range(13)]

fig, ax = plt.subplots()
for i, x in enumerate(data):
    ax.plot(x, label='Legend {}'.format(i))
#ax.legend(loc='upper left')
ax.legend(loc='center left', bbox_to_anchor=(1.0125, 0.5))
```
