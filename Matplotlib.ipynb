```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
mpl.rcParams['font.family'] = 'HGMaruGothicMPRO'

mpl.rcParams['font.size'] = 12
```

# Axes オブジェクト作成する

```python
fig, ax = plt.subplots()
```

# 折れ線グラフ

```python
x = np.random.randn(100)

fig, ax = plt.subplots()
ax.plot(x)
```

```python
fig, ax = plt.subplots()

ax.plot(x, marker='o')
```

```python
fig, ax = plt.subplots()
ax.plot(x, linestyle='dashed')
```

```python
y = np.random.randn(100)

fig, ax = plt.subplots()
ax.plot(x)
ax.plot(y)
```

# 散布図

```python
x = np.random.random(100)
y = np.random.random(100)
s = 500 * np.random.random(100)

fig, ax = plt.subplots()
ax.scatter(x, y, s=s, alpha=0.5)
```

# ヒストグラム

```python
x = np.random.randn(1000)

fig, ax = plt.subplots()
h = ax.hist(x, bins='auto', alpha=0.5)
```

```python
fig, ax = plt.subplots()
h = ax.hist(x, bins='auto', histtype='step')
```
