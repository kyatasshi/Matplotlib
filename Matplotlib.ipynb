# 凡例追加

```python
x = np.random.randn(120)

fig, ax = plt.subplots()
ax.plot(x, label='legend')
ax.legend()
```

```python
x = np.linspace(0, 14, 100)
data = [np.sin(x + i * 0.5) * (13 - i) for i in range(13)]

fig, ax = plt.subplots()
for i, x in enumerate(data):
    ax.plot(x, label='Legend {}'.format(i))
#ax.legend(loc='upper left')
ax.legend(loc='center left', bbox_to_anchor=(1.0125, 0.5))
```

# 可視化ステップ・バイ・ステップ

```python
faang_market_captal = {
    'Facebook': 464190000000,
    'Amazon': 700670000000,
    'Apple': 851320000000,
    'Netflix': 128170000000,
    'Alphabet': 720520000000,
}
faang_share_price = {
    'Facebook': 159.79,
    'Amazon': 1447.34,
    'Apple': 167.78,
    'Netflix': 295.35,
    'Alphabet': 1031.79,
}

names, market_capitals = list(zip(*faang_market_captal.items()))
_, share_prices = list(zip(*faang_share_price.items()))
list(zip(*faang_share_price.items()))
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].barh(names, market_capitals)
ax[1].barh(names, share_prices)
```

```python
fig, ax = plt.subplots(1, 2, sharey=True)
ax[0].barh(names, market_capitals)
ax[1].barh(names, share_prices)
```

```python
fig, axes = plt.subplots(1, 2, sharey=True)
axes[0].barh(names, market_capitals)
axes[1].barh(names, share_prices)

for ax in axes:
    ax.grid()
```

```python
fig, axes = plt.subplots(1, 2, sharey=True)
axes[0].barh(names, market_capitals)
axes[1].barh(names, share_prices)

for ax in axes:
    ax.grid(axis='x', zorder=0)
```

```python
fig, axes = plt.subplots(1, 2, sharey=True)
axes[0].barh(names, market_capitals)
axes[1].barh(names, share_prices)

for ax in axes:
    ax.grid(axis='x', zorder=0)
    
axes[0].set_title('FAANG market capital')
axes[1].set_title('FAANG share price')

axes[0].set_xlabel('Market capital (USD)')
axes[1].set_xlabel('Share price (USD)')
axes[0].set_ylabel('Company')
```
